```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

# Pillow

```python
pic = Image.open('data/00-puppy.jpg')
```

```python
pic
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
print(pic_arr.shape)
print(type(pic_arr))
plt.imshow(X=pic_arr)
```

```python
# Show red channel
plt.imshow(X=pic_arr[:, :, 0], cmap='gray')
```

```python
# Show blue channel
plt.imshow(X=pic_arr[:, :, 1], cmap='gray')
```

```python
# Show green channel
plt.imshow(X=pic_arr[:, :, 2], cmap='gray')
```

```python
# Show red filter
pic_red = pic_arr.copy()
pic_red[:, :, 1:] = 0
plt.imshow(X=pic_red)
```

# OpenCV

```python
img = cv2.imread('data/00-puppy.jpg')
print(type(img))
print(img.shape)
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img_gray = cv2.imread('data/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
print(img_gray.shape)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Resize, width by height
img_rgb_sqz = cv2.resize(img_rgb, (1000, 400))
plt.imshow(img_rgb_sqz)
```

```python
# Resize by ratio
w_ratio = 0.2
h_ratio = 0.2
img_rgb_sqz = cv2.resize(img_rgb, (0, 0), img_rgb, w_ratio, h_ratio)
plt.imshow(img_rgb_sqz)
```

```python
img_rgb_flip = cv2.flip(img_rgb, 1)
plt.imshow(img_rgb_flip)
```

```python
cv2.imwrite(
    'data/rgbpuppy.jpg', 
    cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)
)
```

```python
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(111)
ax.imshow(img_rgb)
```
